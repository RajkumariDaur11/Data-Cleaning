#### The null values are huge, to get an idea of how much our data is having null values let's analyse the % stats.

```python
# Total number of rows and columns

df.shape
```

```python
total_cells = np.product(df.shape)

total_cells
```

```python
total_missing = missing_value_count.sum()

total_missing
```

```python
percent_missing = (total_missing/total_cells)*100
percent_missing
```

#### Observations - 24% of our data is filled with Null Values

#### Analysis - Is this value missing because it wasn't recorded or because it doesn't exist?

If a value it is missing becuase it doesn't exist then it doesn't make sense to try and guess what it might be. 
If a value is missing because it wasn't recorded, then you can try to guess - Imputation

```python
missing_value_count[0:10]
```

```python
# From the documentation the field 'TimeSecs' column has information on the number of seconds left in the game when the play was made. This means that these values are probably missing because they were not recorded.
# Fields like "PenalizedTeam" that also have lot of missing fields. In this case, the field is missing because if there was no penalty then it doesn't make sense to say which team was penalized.
```

### Drop the Nulls

```python
#Remove all the rows that contain a missing value

df.dropna() 
```

```python
# Remove all columns with at least one missing value

columns_with_na_dropped = df.dropna(axis = 1)
columns_with_na_dropped.head(10)
```

```python
# Number of columns dropped

print('Columns in Original Dataset: %d' % df.shape[1])
print('Columns in Dropped Nulls Dataset: %d' % columns_with_na_dropped.shape[1])
```

### Successfully removed all the NaN's from the data !
