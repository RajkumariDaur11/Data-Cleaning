```python
# Importing the libraries

import pandas as pd
import numpy as np

# For Box-Cox Transformations - used where the assumption of normality or constant variance is violated
from scipy import stats
    
# For min-max scaling 
from mlxtend.preprocessing import minmax_scaling

# Plotting Modules
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
```

###### Scaling and Normalization both transforms the values of numeric variables so that the transformed data points have specific helpful properties. 
###### In scaling there is change in range of data and in normalization there is change in the shape of the distribution of data.

# Scaling

```python
# Generate 1000 data points randomly drawn from an exponential distribution

original_data = np.random.exponential(size=1000)

scaled_data = minmax_scaling(original_data, columns=[0])
```

```python
# Plotting and Comparing

fig, ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0], kde=True, legend=False)
ax[0].set_title("Original Data")
sns.histplot(scaled_data, ax=ax[1], kde=True, legend=False)
ax[1].set_title("Scaled Data")
plt.show()
```

```python
# Note - The shape of the data doesn't change, but that instead of ranging from 0 to 8, it now ranges from 0 to 1
```

# Normalization

```python
# Normalize the exponential data with boxcox

normalized_data = stats.boxcox(original_data)
```

```python
# Plotting and comparing

fig, ax  = plt.subplots(1,2, figsize=(15,3))
sns.histplot(original_data, ax=ax[0], legend=False)
ax[0].set_title("Original Data")
sns.histplot(normalized_data, ax=ax[1], legend=False)
ax[1].set_title("Normalized Data")
plt.show()
```
